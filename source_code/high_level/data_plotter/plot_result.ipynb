```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('log.csv', header=None, names=['time','y1','y2','y3','y4','y5','y6'])
range = [18000, 19000]
df = df[range[0]:range[1]]
sns.scatterplot(data=df, x='time', y='y1',label='hip_yaw_angle')
sns.scatterplot(data=df, x='time', y='y2',label='hip_roll_angle')
sns.scatterplot(data=df, x='time', y='y3',label='hip_pitch_angle')
sns.scatterplot(data=df, x='time', y='y4',label='knee_pitch_angle')
sns.scatterplot(data=df, x='time', y='y5',label='ankle_pitch_angle')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Angle (rad)")
plt.show()
```
