```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('log.csv', header=None, names=['time','y1','y2','y3','y4','y5','y6'])
sns.scatterplot(data=df[2100:3400], x='time', y='y1',label='act_knee_angle')
sns.scatterplot(data=df[2100:3400], x='time', y='y2',label='ref_knee_angle')
plt.legend()
plt.xlabel("Time")
plt.ylabel("Angle (rad)")
plt.show()
```
